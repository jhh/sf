# Display a Random Sample Image

Load dependencies:
- [`OpenCV`](http://opencv.org), needs [`numpy`](http://www.numpy.org)
- [`matplotlib`](http://matplotlib.org) used for displaying image
- [`os.path`](https://docs.python.org/3/library/os.path.html#module-os.path) is used for selecting an image to display

```python
%matplotlib inline
import cv2
from matplotlib import pyplot as plt
import numpy as np
import os.path
print("OpenCV Version : %s " % cv2.__version__)
```

We will use OpenCV to open the file and load into a `numpy` array. OpenCV loads the array with pixels in BGR order so we need to convert to RGB for displaying using `matplotlib`.

```python
def display(path):
    image = cv2.imread(path)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.show()
```

Randomly select a sample image and display it.

```python
# there are only 390 sample images, so keep trying random images
# until we get one
path = ""
while not os.path.isfile(path):
    path = "RealFullField/%s.jpg" % np.random.randint(0, 543)

# ok, got one
print(path)
display(path)
```
